```python
#get the data for Agricultural area in establishments with organically
query = Query.region(['110', '11000', '11'])
LF_organic_farms = query.add_field('FLC048')
query.results()
```

```python
#meta data
LF_organic_farms.get_info()
```

```python
#Converted to ecological management
query = Query.region(['110', '11000', '11'])
LF_organic_farms = query.add_field('FLC048')
LF_organic_farms.add_args({'BEWAT3': 'BTRARTOEK05'})
query.results(verbose_enums=True).sort_values('year')
```
